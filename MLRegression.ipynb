## Introduction To Maching Learning
This is the art of making machines/computers intelligent just like humans without explicitly programming.This intelligence will be achieved by making them learn from data.

## Terms:
 1. Artificial Intelligence (AI): AI is basically making computers think/reason and act just like humans. Human intelligence (learning, vision, communicate, motion) is mimiced/copied by machines.

 #### Branches of AI
 a. Machine Learning: Focuses on training computers to learn from historical data and then make future predictions e,g Patients diagnosis, sales forecasting,

 b. Natural Language Processing: Focuses on making computers understand and interpret a natural language (not a programming language e.g speech/text) e,g chatbox, Sentiment Analysis etc.

 c. Robotics: Deals with movements e,g Autonomous Vehicles

 d. Computer Vision: It uses another branch called Deep Learning to make computers understand and see the environment just as humans do. eg Scanning X-Ray Images, Object Detection etc.

```python
#MACHINE LEARNING
# SUPERVISED LEARNING
#STEP 1: Separate the dataset into x & Y
array = df.values
array.shape
```

```python
X= array[:, 0:3]
X.shape
```

```python
Y= array[:, 3]
Y.shape
```

```python
# STEP 2: DIVIDING THE RECORDS(ROWS) INTO TRAINING SETS (70%) AND TESTING SETS(30%)
from sklearn import model_selection
X_train, X_Test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
#STEP3: TRAINING THE MODEL
#TRAINING SETS(X_train,Y_train)
#SCIENTIFIC ALGORITHMS
```

```python
#importing the algorithms
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
```

```python
#STEP 4: CROSS VALIDATION
# DETRMERMINE THE BEST ALGORITHM
```

```python
#training
model = DecisionTreeRegressor()
model.fit(X_train, Y_train)
```

```python
#STEP 5: MODEL EVALUATION: -> Process that determines the accuracy of the model
# ACCURACY
# REGRESSION - r2_score()
# MEAN ABSOLUTE ERROR
```

```python
predictions = model.predict(X_Test)
predictions
```

```python
Y_test
```

```python
#METRICS
from sklearn.metrics import r2_score
accuracy = r2_score(predictions, Y_test) *100
```
